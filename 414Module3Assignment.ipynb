```python
import pandas as pd
import numpy as np
import math
from numpy.linalg import norm
```

```python
league = pd.read_csv('League of Legends pro play Champ Stats from Gol.gg - 12.12.csv')
```

```python
league.sort_values(by='Wins', ascending=False)
```

```python
lol = league.drop(['Champion','Presence','Winrate','KDA','Avg BT','GT','CSD@15','GD@15','XPD@15'],axis = 1)
```

```python
lol
```

```python
lol=lol.fillna(0)
lol
```

```python
Wukong = np.array(lol.iloc[0])
Gnar = np.array(lol.iloc[2])
Aphelios = np.array(lol.iloc[18])
```

```python
league.iloc[0]['Champion']
```

```python
cosine = np.dot(Wukong,Gnar)/(norm(Wukong)*norm(Gnar))
print("Cosine Similarity:", cosine)
```

```python
##Code for Wukong
Wukong_indexed_cosine = pd.DataFrame(columns = ['Champion', 'Cosine Similarity'])
for champ in range(lol.shape[0]):
    comp = np.array(lol.iloc[champ])
    cosine = np.dot(Wukong,comp)/(norm(Wukong)*norm(comp))
    Wukong_indexed_cosine.loc[len(Wukong_indexed_cosine.index)] = [league.iloc[champ]['Champion'],cosine]
    
WukongTop10= Wukong_indexed_cosine.sort_values(by='Cosine Similarity', ascending=False)
WukongTop10.head(10)
```

```python
##Code for Gnar
Gnar_indexed_cosine = pd.DataFrame(columns = ['Champion', 'Cosine Similarity'])
for champ in range(lol.shape[0]):
    comp = np.array(lol.iloc[champ])
    cosine = np.dot(Gnar,comp)/(norm(Gnar)*norm(comp))
    Gnar_indexed_cosine.loc[len(Gnar_indexed_cosine.index)] = [league.iloc[champ]['Champion'],cosine]
    
GnarTop10= Gnar_indexed_cosine.sort_values(by='Cosine Similarity', ascending=False)
GnarTop10.head(10)
```
